#### Activation Function. 
Non Linear function like sigmoid function are called activation function. Activation function is an important part of Artificial Neural Network. They decide whether a neuron should be activated or not. In other words , whether the information recieved by neuron is relevant or should be ignored. 

### Types of Activation function !!

1. Binary Step function
2. Linear function
3. Sigmoid function
4. Hyperbolic Tangent function
5. Rectified Linear unit function
6. Leaky ReLU function
7. Softmax function.

## Common activation functions are 
1. Linear
2. Sigmoid
3. Tanh
4. ReLU
5. ELU

These were great choices as they were differentiable everywhere, monotonic and smooth . However problem such as saturation will occur due to high or low input value to the function. 

```python
Image("pictures/Relu activation function.png")
```

Very popular now is rectifier linear unit or ReLU activation function, It is non linear so we can get the complex modelling needed and it doesnot have the saturation in the non negaitve portion of the input space. However, due to negative portion of the input space  translating to zero activation ReLU layer can end up dying or no longer activating which can also cause training to slow or stop. 

## To solve these problem we can use ELU or exponential Linear Units!

```python
Image("pictures/exponential linear units.png")
```

It is approximately linear and the non-negative portion of the input space, and it's smooth, monotonic, and most importantly non zero in the negative portion of input space. Main drawback of ELU is that they are more computationally expensive than ELU's due to having the calculated exponentials
